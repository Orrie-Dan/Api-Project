```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Mean parameters
mu_x = 0
mu_y = 0

# Standard deviations
sigma_x = 1
sigma_y = 1

# Correlation coefficient
rho = 0.6

# Construct covariance matrix
Sigma = np.array([
    [sigma_x**2, rho * sigma_x * sigma_y],
    [rho * sigma_x * sigma_y, sigma_y**2]
])

# Mean vector
mu = np.array([mu_x, mu_y])

print("Mean Vector (μ):")
print(mu)

print("\nCovariance Matrix (Σ):")
print(Sigma)
```

```python
np.random.seed(42)  # For reproducibility

simulated_data = np.random.multivariate_normal(mu, Sigma, 5000)

X_sim = simulated_data[:, 0]
Y_sim = simulated_data[:, 1]

print("Simulated dataset generated successfully.")
```

```python
def bivariate_normal_pdf(x, y, mu_x, mu_y, sigma_x, sigma_y, rho):
    """
    Computes the probability density function (PDF)
    of a Bivariate Normal Distribution from scratch.
    """

    z_x = (x - mu_x) / sigma_x
    z_y = (y - mu_y) / sigma_y

    exponent = -0.5 / (1 - rho**2) * (z_x**2 - 2*rho*z_x*z_y + z_y**2)
    coefficient = 1 / (2 * np.pi * sigma_x * sigma_y * np.sqrt(1 - rho**2))

    return coefficient * np.exp(exponent)

print("Bivariate Normal PDF function implemented.")
```

```python
x_range = np.linspace(-3, 3, 100)
y_range = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x_range, y_range)

print("Meshgrid created.")
```

```python
Z = bivariate_normal_pdf(X, Y, mu_x, mu_y, sigma_x, sigma_y, rho)

print("PDF values computed.")
```

```python
plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='Probability Density')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of Bivariate Normal Distribution (ρ = 0.6)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Probability Density')
ax.set_title('3D Surface Plot of Bivariate Normal Distribution (ρ = 0.6)')

plt.show()
```
